```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
HH = pd.read_csv('SH_DWH_COM.csv')
HH.head()
```

```python
HH['date'] = pd.to_datetime(HH['date'], format="%d/%m/%Y %H:%M", errors='coerce')
HH = HH.set_index('date')
```

```python
daily_SH_temp_max = HH['sh_temp_mean'].resample('d').max()
daily_SH_temp_min = HH['sh_temp_mean'].resample('d').min()
daily_SH_temp_mean = HH['sh_temp_mean'].resample('d').mean()


daily_SH_hum_max = HH['sh_hum_mean'].resample('d').max()
daily_SH_hum_min = HH['sh_hum_mean'].resample('d').min()
daily_SH_hum_mean = HH['sh_hum_mean'].resample('d').mean()
```

```python
daily_ext_temp_max = HH['Air temp (degC)_mean'].resample('d').max()
daily_ext_temp_min = HH['Air temp (degC)_mean'].resample('d').min()
daily_ext_temp_mean = HH['Air temp (degC)_mean'].resample('d').mean()

daily_ext_RH_max = HH['RH (%)_mean'].resample('d').max()
daily_ext_RH_min = HH['RH (%)_mean'].resample('d').min()
daily_ext_RH_mean = HH['RH (%)_mean'].resample('d').mean()
 
```

```python
daily_Gas_sum = HH['Gas m3_mean'].resample('d').sum()
daily_Gas_mean= HH['Gas m3_mean'].resample('d').mean()
daily_Gas_max = HH['Gas m3_mean'].resample('d').max()
daily_Gas_min = HH['Gas m3_mean'].resample('d').min()
```

```python
daily_stat = pd.DataFrame({
    'Gas_sum': daily_Gas_sum,
    'Gas_max': daily_Gas_max,
    'Gas_min': daily_Gas_min,
    'Gas_mean': daily_Gas_mean,

    
    'ext_temp_max': daily_ext_temp_max,
    'ext_temp_min': daily_ext_temp_min,
    'ext_temp_mean': daily_ext_temp_mean,

    'ext_hum_max': daily_ext_RH_max,
    'ext_hum_min': daily_ext_RH_min,    
    'ext_hum_mean': daily_ext_RH_mean,

    
    'SH_temp_max': daily_SH_temp_max,
    'SH_temp_min': daily_SH_temp_min,    
    'SH_temp_mean': daily_SH_temp_mean,
    
    'SH_hum_max': daily_SH_hum_max,
    'SH_hum_min': daily_SH_hum_min,
    'SH_hum_mean': daily_SH_hum_mean,
    
    'DWH_temp_max': daily_DWH_temp_max,
    'DWH_temp_min': daily_DWH_temp_min,
    'DWH_temp_mean': daily_DWH_temp_mean,
    
    'DWH_hum_max': daily_DWH_hum_max,
    'DWH_hum_min': daily_DWH_hum_min,
    'DWH_hum_mean': daily_DWH_hum_mean,



       

})
```

```python
daily_stat
```

```python
plt.plot(daily_stat['SH_temp_mean'])
plt.plot(daily_stat['ext_temp_mean'])
plt.plot(daily_stat['DWH_temp_mean'])

#plt.plot(daily_stat['Gas_sum'])
```
